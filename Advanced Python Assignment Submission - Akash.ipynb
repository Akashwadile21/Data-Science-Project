 Q1) How has honey production yield changed from 1998 to 2021?

```python
plt.figure(figsize=(14, 7))
sns.barplot(data=df,x='year', y='yieldpercol', ci=None,edgecolor='black')
plt.xticks(rotation=90) 
plt.title("Honey Production Yield  from 1998 to 2021")
plt.xlabel('Year',fontsize=13)
plt.ylabel('Yield Production',fontsize=13)

plt.show()
```

Q2) Over time, what are the major production trends across the
states?

```python
plt.figure(figsize=(14, 7))
sns.scatterplot(data=df,x='year', y='totalprod', hue = 'State')
plt.legend(ncol=2, fontsize=8,markerscale=0.7, bbox_to_anchor=(1,1))
plt.title('Major Production Across the States')
plt.xlabel('Year',fontsize=13)
plt.ylabel('Total production',fontsize=13)
plt.xticks(rotation=90) 
plt.show()
```

Q3) Does the data show any trends in terms of the number of
honey-producing colonies and yield per colony before 2006, which
was when concern over Colony Collapse Disorder spread nationwide?


```python
data=df[df["year"].isin([1998,1999,2000,2001,2002,2003,2004,2005,2006])]
data
```

```python
total_colonies=data.groupby('year')['numcol'].sum().head(9)
total_colonies
```

```python
yield_colonies=data.groupby('year')['yieldpercol'].sum().head(9)
yield_colonies
```

```python

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(15,11))


ax1.plot(total_colonies.index,total_colonies.values,marker="o",color="orange",markersize=7)
ax1.set_title('Number of Honey-Producing Colonies (Before 2006)',color='red')
ax1.set_xlabel('Year')
ax1.set_ylabel('Number of Colonies')


ax2.plot(yield_colonies.index,yield_colonies.values,marker="o",markersize=7)
ax2.set_title('Yield Per Colony (Before 2006)')
ax2.set_xlabel('Year')
ax2.set_ylabel('Yield per Colony')

plt.show()
```

Q4)Are there any patterns that can be observed between total honey
production and the value of production every year?

```python
Honey_production=df.groupby('year')['totalprod'].sum()
Honey_production
```

```python
production_value=df.groupby('year')['prodvalue'].sum()
production_value
```

```python

fig, (ax1, ax2) = plt.subplots(2,1,figsize=(15, 11))

ax1.bar(Honey_production.index,Honey_production.values,color='orchid',edgecolor='black')
ax1.set_title('Total Hooney Production Every Year')
ax1.set_xlabel('Year')
ax1.set_ylabel('Total Production')


ax2.bar(production_value.index,production_value.values,color='orange',edgecolor='black')
ax2.set_title('Production Value Every Year')
ax2.set_xlabel('Year')
ax2.set_ylabel('Value of Production')

plt.show()
```
