```python
def plot_topN_col(feat,N):
  res=df_spotify[feat].value_counts().head(N)
#  return res
  plt.barh(res.index,res.values, color='orange')
  plt.title(f'Top {N} Artist with most songs in top 100 spotify song')
  plt.grid()
  plt.show()
```

```python
plot_topN_col('artist',7)
```

```python
def plot_topN_col(feat,N,featname):
  res=df_spotify[feat].value_counts().head(N)
#  return res
  plt.barh(res.index,res.values, color='orange')
  plt.title(f'Top {N}{featname} with most songs in top 100 spotify song')
  plt.grid()
  plt.show()
```

```python
plot_topN_col('artist',7,'Artist')
```

```python
plot_topN_col('top genre',11,'Gennre')
```
